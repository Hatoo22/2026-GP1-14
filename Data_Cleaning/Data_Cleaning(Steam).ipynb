# 3) Analyze language distribution


This step analyzes the distribution of languages in the Steam reviews dataset. The data is loaded in chunks to handle large file sizes efficiently without consuming too much memory.

For each chunk, the number of reviews per language is counted using the "language" column. These counts are then aggregated to compute the total number of reviews for each language across the entire dataset.

Finally, the code calculates and displays the percentage of each language relative to the total number of reviews, providing a clear overview of language distribution in the dataset.


```
# =========================
# 3) Analyze language distribution
# =========================
import pandas as pd

file = "/content/drive/MyDrive/steam_reviews.csv"

lang_counts = {}
total_rows = 0

for chunk in pd.read_csv(file, chunksize=5000000):

    total_rows += len(chunk)

    counts = chunk["language"].value_counts()

    for lang, c in counts.items():
        lang_counts[lang] = lang_counts.get(lang, 0) + c

print("عدد الصفوف الكلي:", total_rows)
print("\nتوزيع اللغات:\n")

for lang, count in sorted(lang_counts.items(), key=lambda x: x[1], reverse=True):
    percentage = (count / total_rows) * 100
    print(f"{lang}: {count} ({percentage:.2f}%)")
```

# 4) Filter English reviews only


This step filters the dataset to keep only English reviews. The data is processed in chunks to efficiently handle large file sizes. For each chunk, rows where the language is labeled as "english" are selected and saved into a new file. The result is a separate dataset containing only English reviews.

# 8) Extract k rows per game


In this step, we define a function that extracts the first k (10 k) rows for each unique group in a specified column (app_id). This ensures that the dataset contains a balanced number of reviews from each game, which helps maintain diversity and prevents the dataset from being dominated by reviews from a single game.

```
# =========================
# 8) Extract k rows per game
# =========================
import pandas as pd

def get_first_k_rows_per_group(df, group_col, k):
    df_list = []
    unique_groups = df[group_col].unique()
    for group_val in unique_groups:
        filtered_df = df[df[group_col] == group_val]
        df_list.append(filtered_df.head(k))
    return pd.concat(df_list, ignore_index=True)

k_rows_per_app_id = get_first_k_rows_per_group(df, 'app_id', 10000)

display(k_rows_per_app_id.head())
display(f"Total rows in the new DataFrame: {len(k_rows_per_app_id)}")
```

# **Data Cleaning**
In this block, basic cleaning steps were applied to prepare the data for further processing. First, rows with missing values in the review column were removed, as empty reviews do not provide useful information for analysis or model training. Then, all values in the review column were converted into string format to ensure consistency when handling text data. In addition, other cleaning steps were applied to improve data quality. Finally, the cleaned dataset was saved into a new file to be used later in model training and evaluation.

# 11) Data CLEANING


```
# =========================
# 11) Data CLEANING
# =========================

# Remove missing values
df = df.dropna(subset=['review'])

# Convert everything to text
df['review'] = df['review'].astype(str)

# Remove extra spaces from the beginning and end
df['review'] = df['review'].str.strip()

# Remove empty rows or rows that contain only spaces.
df = df[df['review'] != ""]

# Normalize spaces within the text
df['review'] = df['review'].str.replace(r'\s+', ' ', regex=True)

# Remove rows that contain only numbers
df = df[~df['review'].str.match(r'^\d+$', na=False)]

# Remove rows that contain only symbols
df = df[~df['review'].str.match(r'^[^\w\s]+$', na=False)]

# Remove rows that do not contain any English letters
df = df[df['review'].str.contains(r'[a-zA-Z]', regex=True, na=False)]

# Remove very short rows (less than 3 words)
df = df[df['review'].str.split().str.len() >= 3]

# Remove very short rows based on character length
df = df[df['review'].str.len() >= 8]

# Remove duplicates
df = df.drop_duplicates(subset=['review']).reset_index(drop=True)

print("عدد التعليقات بعد التنظيف:", len(df))
print(df.head())

# Save the cleaned dataset
df.to_csv('clean_reviews.csv', index=False, encoding='utf-8-sig')
print("تم حفظ الملف: clean_reviews.csv")
```
