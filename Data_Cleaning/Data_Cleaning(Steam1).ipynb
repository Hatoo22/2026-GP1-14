# **Data Loading and Inspection**
In this block, the dataset was loaded from a CSV file using the pandas library. Since the file is very large (over 400,000 reviews), we faced inconsistencies when reading it with default settings. To handle this, we specified the Python engine and defined the separator to ensure the file is read correctly without skipping or misinterpreting rows. After loading the data, we selected and retained only the review column, as it is the main feature needed for our text analysis task, while unnecessary columns were excluded. This step ensures that all records are properly loaded and that the dataset is focused and ready for the cleaning phase.

```
import pandas as pd
import re

# =========================
# 1) READ CSV CORRECTLY
# =========================
df = pd.read_csv(
    'k_rows_per_app_id.csv',
    engine='python',
    sep=',',
    quotechar='"',
    escapechar='\\',
    on_bad_lines='skip'
)

print("Shape before selecting columns:", df.shape)

# We keep only the review column
df = df[['review']].copy()

print("عدد التعليقات قبل التنظيف:", len(df))
```

# **Data Cleaning**
In this block, basic cleaning steps were applied to prepare the data for further processing. First, rows with missing values in the review column were removed, as empty reviews do not provide useful information for analysis or model training. Then, all values in the review column were converted into string format to ensure consistency when handling text data. In addition, other cleaning steps were applied to improve data quality. Finally, the cleaned dataset was saved into a new file to be used later in model training and evaluation.

```
# =========================
# 2) Data CLEANING
# =========================

# Remove missing values
df = df.dropna(subset=['review'])

# Convert everything to text
df['review'] = df['review'].astype(str)

# Remove extra spaces from the beginning and end
df['review'] = df['review'].str.strip()

# Remove empty rows or rows that contain only spaces.
df = df[df['review'] != ""]

# Normalize spaces within the text
df['review'] = df['review'].str.replace(r'\s+', ' ', regex=True)

# Remove rows that contain only numbers
df = df[~df['review'].str.match(r'^\d+$', na=False)]

# Remove rows that contain only symbols
df = df[~df['review'].str.match(r'^[^\w\s]+$', na=False)]

# Remove rows that do not contain any English letters
df = df[df['review'].str.contains(r'[a-zA-Z]', regex=True, na=False)]

# Remove very short rows (less than 3 words)
df = df[df['review'].str.split().str.len() >= 3]

# Remove very short rows based on character length
df = df[df['review'].str.len() >= 8]

# Remove duplicates
df = df.drop_duplicates(subset=['review']).reset_index(drop=True)

print("عدد التعليقات بعد التنظيف:", len(df))
print(df.head())

# Save the cleaned dataset
df.to_csv('clean_reviews.csv', index=False, encoding='utf-8-sig')
print("تم حفظ الملف: clean_reviews.csv")
```
